## Nhận xét về tính kịp thời của MOOCCubeX

Bộ dữ liệu MOOCCubeX thiếu đi thuộc tính thể hiện sự thay đổi theo thời gian. Tuy nhiên, một số dấu hiệu cho thấy khả năng cập nhật liên tục của MOOCCubeX, đặc biệt trong phần "Video Watching Behavior". Cụ thể:
- *Dữ liệu được thu thập trong một khoảng thời gian nhất định:* Nhóm tác giả thu thập dữ liệu học tập qua video trong thời gian 19/01/2020 đến 03/11/2020, gồm 4124 khóa học.
- *Lượng dữ liệu khổng lồ:* 154.332.174 dữ liệu thô về nhật ký học tập được ghi lại, với tần suất 5 giây một lần, ghi lại lịch sử học tập của người dùng (video đang xem và vị trí).
- *Mức độ chi tiết cao:* Dữ liệu được lưu trữ chi tiết trong tập tin "relations/user-video.json". Phân tích 1000 thông tin đầu tiên cho thấy sự tỉ mỉ trong việc ghi chép hành vi học tập của người dùng.
Vậy nên, ta có thể ghi nhận rằng MOOCCubeX có khả năng cập nhật liên tục, đặc biệt là phần dữ liệu liên quan đến hành vi xem video. Tuy nhiên, để khẳng định chắc chắn, cần có thêm thông tin về quy trình cập nhật dữ liệu của nhóm nghiên cứu.

```python
user_video = pd.read_json(FILES['user-video'], lines=True, nrows=1000)
user_video
```

Mỗi người dùng sẽ xem nhiều video, mỗi video có các mã định danh (video_id) riêng biệt. Các video này được chia thành các phân đoạn (segment), bao gồm các phần nhỏ khác nhau. Mỗi phần nhỏ thể hiện thời điểm cụ thể trong video và tốc độ phát lại tương ứng.

```python
user_video['seq'][0]
```

```python
user_video['seq'][0][0]['segment']
```

*Nhận xét:*
- local_start_time: Thời gian bắt đầu của mỗi phần nhỏ trong video.
- (end_point - start_point)/speed: Khoảng thời gian trong thời điểm video đang xét, được tính bằng cách chia hiệu số giữa điểm kết thúc (end_point) và điểm bắt đầu (start_point) cho tốc độ phát lại (speed).
Vì local_start_time không đồng nhất và không khớp với (end_point - start_point)/speed, nên có thể kết luận rằng: *Dữ liệu đã được xử lý để loại bỏ thông tin trùng lặp.*

Lý do:
- Khi người dùng tạm dừng hoặc tua video, dữ liệu về khoảng thời gian đó sẽ bị trùng lặp.
- Để giảm dung lượng lưu trữ và tăng hiệu quả xử lý, dữ liệu trùng lặp sẽ được loại bỏ.
- Việc loại bỏ dữ liệu trùng lặp khiến cho local_start_time không đồng nhất và không khớp với (end_point - start_point)/speed.

=> Việc kiểm tra tính kịp thời trong trường hợp này gặp nhiều khó khăn.
